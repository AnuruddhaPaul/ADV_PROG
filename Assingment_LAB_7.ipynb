```python

bin_start = 50000
bin_increment = 15000
max_salary = df['salary'].max()
bins = []
current = bin_start
while current <= max_salary + bin_increment:
    bins.append(current)
    current += bin_increment

print(f"\nSalary Range: ${df['salary'].min():,.2f} - ${df['salary'].max():,.2f}")
print(f"Bins: {bins}")


plt.figure(figsize=(12, 6))
n, bins_used, _ = plt.hist(df['salary'], bins=bins, edgecolor='black',
                                  color='skyblue', alpha=0.7)

plt.xlabel('Salary ($)', fontsize=12, fontweight='bold')
plt.ylabel('Number of Faculty', fontsize=12, fontweight='bold')
plt.title('Histogram of Faculty Salaries',
          fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)


ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

for i in range(len(n)):
    if n[i] > 0:
        plt.text(bins_used[i] + bin_increment/2, n[i], str(int(n[i])),
                ha='center', va='bottom', fontweight='bold')


plt.show()


```
