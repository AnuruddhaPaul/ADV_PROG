```python

n = int(input("Enter the number of terms: "))
# First two terms
a, b = 0, 1
series = []

if n == 1:
    series.append(a)
else:
    series.append(a)
    series.append(b)
    # Generate remaining terms
    for _ in range(2, n):
        next_term = a + b
        series.append(next_term)
        a, b = b, next_term
        
print(f"Fibonacci series ({n} terms): {series}")
```

```python

num = int(input("Enter a number: "))
original_num = num
factors = set()  


while num % 2 == 0:
    factors.add(2)
    num = num // 2
    

divisor = 3
while divisor * divisor <= num:
    while num % divisor == 0:
        factors.add(divisor)
        num = num // divisor
    divisor += 2
    

print(factors)
if num > 2:
    factors.add(num)
    

sorted_factors = sorted(list(factors))
print(f"Distinct prime factors of {original_num}: {sorted_factors}")

```

```python
import random


target_number = random.randint(1, 100)
attempts = 0
guessed_correctly = False

print("--- Number Guessing Game ---")
print("I have picked a number between 1 and 100.")

while not guessed_correctly:
    
    user_input = input("Guess the number (or 'q' to quit): ")
    
    
    if user_input.lower() == 'q':
        print(f"Quitting... The number was {target_number}.")
        break
    
    guess = int(user_input)
    attempts += 1
    
    
    if guess == target_number:
        print(f"Congratulations! You guessed it in {attempts} attempts.")
        guessed_correctly = True
    elif guess < target_number:
        print("Try Higher!")
    else:
        print("Try Lower!")
        
    

```
