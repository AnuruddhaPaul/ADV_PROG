```python
class Holiday():
  def __init__(self, name, day, month):
    self.name = name
    self.day = day
    self.month = month
  def inSameMonth(self, other):
    return self.month == other.month
  def avgDate(holidays):
    if not holidays:
        return 0.0
    total = sum(holiday.day for holiday in holidays)
    return total / len(holidays)

independence_day = Holiday("Independence Day", 15, "August")
christmas_day = Holiday("Christmas Day", 25, "December")
print(independence_day.inSameMonth(christmas_day))
print(Holiday.avgDate([independence_day, christmas_day]))
```

```python
class Movie:
    def __init__(self, title, studio, rating="PG"):
        self.title = title
        self.studio = studio
        self.rating = rating


    def getPG(movies):
        pg_movies = []
        for movie in movies:
            if movie.rating == "PG":
                pg_movies.append(movie)
        return pg_movies

    def __str__(self):
        return f"\'{self.title}\' by {self.studio} (Rated: {self.rating})"


casino_royale = Movie("Casino Royale", "Eon Productions", "PG13")

if __name__ == "__main__":
    print("Created Movie:")
    print(casino_royale)



    movie1 = Movie("The Lion King", "Disney", "PG")
    movie2 = Movie("The Dark Knight", "Warner Bros", "PG13")
    movie3 = Movie("Toy Story", "Pixar")
    movie4 = Movie("Inception", "Warner Bros", "PG13")
    movie5 = Movie("Finding Nemo", "Pixar", "PG")
    movie6 = Movie("Frozen", "Disney")

    all_movies = [movie1, movie2, movie3, movie4, movie5, movie6]

    print("\nAll Movies:")
    for movie in all_movies:
        print(f"  {movie}")

    pg_movies = Movie.getPG(all_movies)

    print(f"\nPG Rated Movies (Found {len(pg_movies)}):")
    for movie in pg_movies:
        print(f"  {movie}")
```

```python


class Employee:


    companyName = "TechCorp Solutions"

    def __init__(self, name, age, experience, department, salary):

        self.name = name
        self.__age = age
        self.experience = experience
        self.department = department
        self.__salary = salary


    def get_age(self):

        return self.__age

    def get_salary(self):

        return self.__salary


    def set_age(self, age):

        if age > 0:
            self.__age = age

    def set_salary(self, salary):

        if salary > 0:
            self.__salary = salary

    def display_info(self):

        print(f"\nEmployee: {self.name}")
        print(f"Company: {Employee.companyName}")
        print(f"Age: {self.__age}")
        print(f"Experience: {self.experience} years")
        print(f"Department: {self.department}")
        print(f"Salary: ${self.__salary:,.2f}")

    def get_annual_bonus(self):

        return self.__salary * 0.10

    def promote(self, salary_increase):

        self.__salary += salary_increase
        print(f"{self.name} promoted! New salary: ${self.__salary:,.2f}")

    def __str__(self):
        return f"{self.name} - {self.department} ({self.experience} yrs exp)"


class Manager(Employee):


    def __init__(self, name, age, experience, department, salary, teams=None):

        super().__init__(name, age, experience, department, salary)
        self.teams = teams if teams is not None else []

    def add_team(self, team_name):

        if team_name not in self.teams:
            self.teams.append(team_name)
            print(f"Team \'{team_name}\' assigned to {self.name}")
        else:
            print(f"Team \'{team_name}\' already supervised by {self.name}")

    def remove_team(self, team_name):

        if team_name in self.teams:
            self.teams.remove(team_name)
            print(f"Team \'{team_name}\' removed from {self.name}\'s supervision")
        else:
            print(f"Team \'{team_name}\' not found")

    def get_total_teams(self):

        return len(self.teams)

    def display_info(self):

        super().display_info()
        print(f"Role: Manager")
        print(f"Teams Managed: {self.get_total_teams()}")
        if self.teams:
            print(f"Team Names: {', '.join(self.teams)}")

    def conduct_meeting(self):

        print(f"{self.name} is conducting a meeting with {self.get_total_teams()} team(s)")

    def __str__(self):
        return f"Manager {self.name} - {self.department} (Managing {self.get_total_teams()} teams)"


class Developer(Employee):

    def __init__(self, name, age, experience, department, salary,
                 programming_languages=None, projects=None, specialization="Full Stack"):

        super().__init__(name, age, experience, department, salary)
        self.programming_languages = programming_languages if programming_languages is not None else []
        self.projects = projects if projects is not None else []
        self.specialization = specialization
        self.lines_of_code = 0

    def add_programming_language(self, language):

        if language not in self.programming_languages:
            self.programming_languages.append(language)
            print(f"{self.name} learned {language}")

    def add_project(self, project_name):

        if project_name not in self.projects:
            self.projects.append(project_name)
            print(f"Project \'{project_name}\' assigned to {self.name}")

    def write_code(self, lines):

        self.lines_of_code += lines
        print(f"{self.name} wrote {lines} lines of code. Total: {self.lines_of_code}")

    def debug_code(self):

        print(f"{self.name} is debugging code...")

    def code_review(self, reviewer_name):

        print(f"{self.name}\'s code is being reviewed by {reviewer_name}")

    def get_project_count(self):

        return len(self.projects)

    def display_info(self):

        super().display_info()
        print(f"Role: Software Developer")
        print(f"Specialization: {self.specialization}")
        print(f"Programming Languages: {', '.join(self.programming_languages) if self.programming_languages else 'None'}")
        print(f"Projects: {self.get_project_count()}")
        print(f"Total Lines of Code: {self.lines_of_code:,}")

    def __str__(self):
        return f"Developer {self.name} - {self.specialization} ({len(self.programming_languages)} languages)"




if __name__ == "__main__":

    print("EMPLOYEE HIERARCHY DEMONSTRATION")
    print("=" * 60)


    print(f"\nCompany: {Employee.companyName}")
    print("\n--- Creating Employees ---")

    emp1 = Employee("Rajesh Kumar", 28, 3, "Finance", 55000)
    print(f"Created: {emp1}")

    manager1 = Manager("Priya Sharma", 35, 10, "Engineering", 95000,
                       ["Backend Team", "DevOps Team"])
    print(f"Created: {manager1}")

    dev1 = Developer("Amit Patel", 26, 4, "Engineering", 75000,
                     ["Python", "Java", "JavaScript"],
                     ["E-commerce Platform", "Payment Gateway"],
                     "Backend Developer")
    print(f"Created: {dev1}")

    dev2 = Developer("Sneha Reddy", 24, 2, "Engineering", 65000,
                     ["React", "TypeScript", "CSS"],
                     ["Mobile App"],
                     "Frontend Developer")
    print(f"Created: {dev2}")


    print("\n--- Manager Operations ---")
    manager1.add_team("Mobile Team")
    print(f"Total teams managed by {manager1.name}: {manager1.get_total_teams()}")
    manager1.conduct_meeting()


    print("\n--- Developer Operations ---")
    dev1.write_code(500)
    dev1.add_programming_language("Go")
    dev1.add_project("Microservices Architecture")
    dev1.debug_code()


    print("\n--- Employee Details ---")
    emp1.display_info()
    manager1.display_info()
    dev1.display_info()


```

```python

class TreeNode:


    def __init__(self, value):

        self.value = value
        self.left = None
        self.right = None

    def __str__(self):
        return str(self.value)


class BinaryTree:


    def __init__(self, root_value=None):

        self.root = TreeNode(root_value) if root_value is not None else None

    def get_height(self, node=None, _first_call=True):

        if _first_call and node is None:
            node = self.root


        if node is None:
            return 0


        left_height = self.get_height(node.left, False)
        right_height = self.get_height(node.right, False)


        return max(left_height, right_height) + 1

    def count_nodes(self, node=None, _first_call=True):

        if _first_call and node is None:
            node = self.root


        if node is None:
            return 0


        return 1 + self.count_nodes(node.left, False) + self.count_nodes(node.right, False)

    def add_node(self, value):

        new_node = TreeNode(value)

        if self.root is None:
            self.root = new_node
            print(f"Added {value} as root node")
            return


        queue = [self.root]

        while queue:
            current = queue.pop(0)


            if current.left is None:
                current.left = new_node
                print(f"Added {value} as left child of {current.value}")
                return
            else:
                queue.append(current.left)


            if current.right is None:
                current.right = new_node
                print(f"Added {value} as right child of {current.value}")
                return
            else:
                queue.append(current.right)

    def remove_node(self, value):

        if self.root is None:
            print("Tree is empty")
            return False


        if self.root.left is None and self.root.right is None:
            if self.root.value == value:
                self.root = None
                print(f"Removed node {value} (was root)")
                return True
            else:
                print(f"Node {value} not found")
                return False


        queue = [self.root]
        target_node = None
        last_node = None
        parent_of_last = None

        while queue:
            current = queue.pop(0)

            if current.value == value:
                target_node = current

            if current.left:
                parent_of_last = current
                last_node = current.left
                queue.append(current.left)

            if current.right:
                parent_of_last = current
                last_node = current.right
                queue.append(current.right)

        if target_node is None:
            print(f"Node {value} not found")
            return False


        target_value = target_node.value
        target_node.value = last_node.value

        if parent_of_last.right == last_node:
            parent_of_last.right = None
        else:
            parent_of_last.left = None

        print(f"Removed node {target_value}")
        return True

    def inorder_traversal(self, node=None, result=None, _first_call=True):

        if result is None:
            result = []

        if _first_call and node is None:
            node = self.root

        if node is None:
            return result

        if node.left:
            self.inorder_traversal(node.left, result, False)

        result.append(node.value)

        if node.right:
            self.inorder_traversal(node.right, result, False)

        return result

    def preorder_traversal(self, node=None, result=None, _first_call=True):

        if result is None:
            result = []

        if _first_call and node is None:
            node = self.root

        if node is None:
            return result

        result.append(node.value)

        if node.left:
            self.preorder_traversal(node.left, result, False)

        if node.right:
            self.preorder_traversal(node.right, result, False)

        return result

    def level_order_traversal(self):

        if self.root is None:
            return []

        result = []
        queue = [self.root]

        while queue:
            current = queue.pop(0)
            result.append(current.value)

            if current.left:
                queue.append(current.left)
            if current.right:
                queue.append(current.right)

        return result

    def display(self):

        if self.root is None:
            print("Tree is empty")
            return

        print("\nTree Structure:")
        print(f"  Root: {self.root.value}")
        print(f"  Height: {self.get_height()}")
        print(f"  Total Nodes: {self.count_nodes()}")
        print(f"  Level-order: {self.level_order_traversal()}")
        print(f"  Inorder: {self.inorder_traversal()}")
        print(f"  Preorder: {self.preorder_traversal()}")


# ============================================
# Example Usage
# ============================================

if __name__ == "__main__":

    print("BINARY TREE DEMONSTRATION")
    print("=" * 60)


    bt = BinaryTree(1)
    print("\n--- Creating Binary Tree ---")
    print("Root node created with value: 1")

    print("\n--- Adding Nodes ---")
    bt.add_node(2)
    bt.add_node(3)
    bt.add_node(4)
    bt.add_node(5)
    bt.add_node(6)
    bt.add_node(7)


    bt.display()


    print("\n--- Adding More Nodes ---")
    bt.add_node(8)
    bt.add_node(9)
    bt.display()


    print("\n--- Removing Nodes ---")
    bt.remove_node(5)
    bt.display()

```

```python
class Node:
    def __init__(self, v): self.val, self.left, self.right = v, None, None

class BinaryTree:
    def __init__(self): self.root = None

    def add(self, val):
        if not self.root: self.root = Node(val); return
        q = [self.root]
        while q:
            curr = q.pop(0)
            if not curr.left: curr.left = Node(val); return
            if not curr.right: curr.right = Node(val); return
            q.extend([curr.left, curr.right])

    def inorder(self, n):
        return self.inorder(n.left) + [n.val] + self.inorder(n.right) if n else []

    def height(self, n):
        return 1 + max(self.height(n.left), self.height(n.right)) if n else 0

if __name__ == "__main__":
    bt = BinaryTree()
    for i in range(1, 8): bt.add(i)  # Adds 1 through 7

    print(f"Height: {bt.height(bt.root)}")
    print(f"Inorder: {bt.inorder(bt.root)}")
```
