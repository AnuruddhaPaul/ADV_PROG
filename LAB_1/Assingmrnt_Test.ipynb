```python

print("="*70)
print("Q.1) Variables, Types, Delete, and Comparison")
print("="*70)


int_var = 42
float_var = 3.14
str_var = "Wildlife Detection"
bool_var = True
list_var = [1, 2, 3]
dict_var = {'arch': 'YOLO-HARVEST', 'year': 2025}


print("\nVariables and their values:")
print("int_var:", int_var)
print("float_var:", float_var)
print("str_var:", str_var)
print("bool_var:", bool_var)
print("list_var:", list_var)
print("dict_var:", dict_var)

print("\nTypes of variables:")
print("type(int_var):", type(int_var))
print("type(float_var):", type(float_var))
print("type(str_var):", type(str_var))
print("type(bool_var):", type(bool_var))
print("type(list_var):", type(list_var))
print("type(dict_var):", type(dict_var))


print("\nDeleting bool_var...")
del bool_var
print("bool_var has been deleted")


num1 = 7
num2 = 10
print("\nInteger comparison:")
print(f"num1 = {num1}")
print(f"num2 = {num2}")
print(f"num1 > num2: {num1 > num2}")
print(f"num1 < num2: {num1 < num2}")
print(f"num1 == num2: {num1 == num2}")
```

```python
print("\n" + "="*70)
print("Q.3) Find biggest of two numbers (using operators only)")
print("="*70)

a = 25
b = 40
biggest = (a + b + abs(a - b)) // 2
print(f"Number 1: {a}")
print(f"Number 2: {b}")
print(f"Biggest number: {biggest}")
```
